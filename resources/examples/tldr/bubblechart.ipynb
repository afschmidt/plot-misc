# Bubble chart

A bubble chart is essentially a heatmap using dots instead of squares. These charts often combine colour and size aesthetics to convey meaning.

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

import plot_misc.example_data.examples as examples
from plot_misc import incidencematrix as imat_plt

CMTOINCH = 1 / 2.54

# get data
table, DOTS, handles = examples.load_bubble_preprocessed()

# creating a plot with room for a top legend
fig = plt.figure(figsize=(4 * CMTOINCH, 6 * CMTOINCH))
gs = GridSpec(nrows=2, ncols=1, height_ratios=[1, 10], wspace=0.1, figure=fig)
ax = fig.add_subplot(gs[1])
leg_ax = fig.add_subplot(gs[0])

# plotting
_ = imat_plt.draw_incidencematrix(
    data=table,
    ax=ax,
    lw=[1, 1],
    line_colour=["black", "black"],
    dot_colour=DOTS[0],
    dot_size=DOTS[1],
    dot_transparency=[1],
    tick_lab_size=[10, 10],
    tick_len=[0, 0],
    margins=[0, 0],
    grid_position="outline",
    kwargs_scatter_dict={"edgecolor": "black", "linewidths": 0.4},
)

# legend
leg_ax.axis("off")
_ = leg_ax.legend(
    handles=[*handles],
    loc="center",
    bbox_to_anchor=(0.5, -0.4),
    ncol=len(handles),
    fontsize=7,
    frameon=False,
    handletextpad=0.3,
    columnspacing=0.3,
)
```
