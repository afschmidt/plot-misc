# Incidence matrix

Incidence-matrix plots represent binary tables as dot grids, where colour and size encode category membership.

```python
import numpy as np

import plot_misc.example_data.examples as examples
from plot_misc import incidencematrix as imat_plt

CMTOINCH = 1 / 2.54

# get data
data = examples.get_data("load_incidence_matrix_data").head(10)
del data["Counts"]

# Plotting
_ = imat_plt.draw_incidencematrix(
    data,
    fsize=(5 * CMTOINCH, 6 * CMTOINCH),
    dot_colour=[("#C4C4C4", 0), ("#B12137", np.inf)],
    dot_size=[20, 70],
    tick_lab_size=(8, 8),
    tick_len=[0, 0],
    lw=[0.5, 0.5],
    line_colour=["black", "black"],
    margins=[0.05, 0.05],
    kwargs_scatter_dict={
        "edgecolor": "black",
        "linewidths": 0.8,
    },
)
```
